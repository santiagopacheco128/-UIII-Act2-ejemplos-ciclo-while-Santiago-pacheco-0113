```python
# Ejemplo 1
contador = 1

while contador <= 5:
    print(contador)
    contador += 1
```

```python
# Ejemplo 2
numero = 10

while numero >= 1:
    print(numero)
    numero -= 1
```

```python
# Ejemplo 3
respuesta = ""

while respuesta != "si":
    respuesta = input("Escribe si para salir: ")
```

```python
# Ejemplo 4
numero = 5
contador = 1

while contador <= 10:
    print(numero, "x", contador, "=", numero * contador)
    contador += 1
```

```python
# Ejemplo 5
suma = 0
contador = 1

while contador <= 5:
    suma += contador
    contador += 1

print("La suma es:", suma)
```
